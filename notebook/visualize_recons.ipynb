```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

base_path = '/mnt/disk1/heonseok/MPMLD/output0727/SVHN/reconstructor/'
if not os.path.exists('Figs'):
    os.mkdir('Figs')
```

```python
def merge_recons(model_path):
    recon_type_list = [
       'cb_mb.png',
       'cz_mb.png',
       'cb_mz.png',
    ]

    plt.figure(1, figsize=(20,8))

    for recon_idx, recon_type in enumerate(recon_type_list):
        plt.subplot(str(13)+str(recon_idx+1))
        plt.imshow(mpimg.imread(os.path.join(model_path, recon_type)))
        plt.xticks([])
        plt.yticks([])

    plt.show()
```
